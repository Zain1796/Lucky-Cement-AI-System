```python
import pandas as pd
import json

# Load Excel file
file_path = "/content/drive/MyDrive/Joyn_NewDataNLP/Excel_sheets/WhisperTrainData.xlsx"
xls = pd.ExcelFile(file_path)
# exclude = ["pre-finetuned-allDSR", "Oct_21_2023", "Nov_18_2023",  "Nov_20_2023"]
exclude = xls.sheet_names[-18:]
# Iterate over each sheet and process
all_data = []
for sheet_name in xls.sheet_names:
    if sheet_name not in exclude:
      df = pd.read_excel(xls, sheet_name=sheet_name)

      # Find the column immediately before 'Corrected'
      corrected_index = df.columns.get_loc("Corrected")
      transcript_column = df.columns[corrected_index - 1]
      if "Transcript_D" in transcript_column:
        transcript_column = df.columns[corrected_index]
      # Convert the columns to string to avoid 'float' type issues
      print(transcript_column)
      df[transcript_column] = df[transcript_column].fillna('')
      df['Corrected'] = df['Corrected'].fillna('-??-')
      # Merge using 'Corrected' if available, otherwise use the identified column
      df['To_Merge'] = df.apply(lambda row: row['Corrected'] if row['Corrected'] else row[transcript_column], axis=1)
      merged_transcript = df.groupby(df['Audio Name'].str.rsplit('_', 1).str[0])['To_Merge'].apply(' '.join).reset_index()
      all_data.append(merged_transcript)

# Combine all sheets data
combined_data = pd.concat(all_data)
combined_data.columns = ['Audio Name', 'Merged Transcript']

# Save the combined transcripts to an Excel file
output_path = "combined_transcripts.xlsx"
combined_data.to_excel(output_path, index=False)

# Create a JSON structure
json_data = [{"text": transcript, "label": []} for transcript in combined_data["Merged Transcript"]]

# Save to a JSONL file
output_jsonl_path = "combined_transcripts.jsonl"
with open(output_jsonl_path, 'w', encoding='utf-8') as jsonl_file:
    for item in json_data:
        jsonl_file.write(json.dumps(item, ensure_ascii=False) + '\n')
```
