This Coede will Replace the brand in Text and Label

```python
import json
import re
import codecs
import random

def replace_text(text, old_word, new_word):
    return text.replace(old_word, new_word)


def replace_label(label, old_brand, new_brand):
    return [[item[0], item[1], item[2].replace(old_brand, new_brand)] for item in label]


file_path = '/content/drive/MyDrive/Joyn_NewDataNLP/NER/all_NER_iter1.jsonl'
output_path = ''
old_word = 'لکی'#input('یہاں جو لفظ آپ تبدیل کرنا چاہتے ہیں  :')
new_word = 'پاور' #input('یہاں جس  لفظ سے آپ تبدیل کرنا چاہتے ہیں  :')
old_brand = "Lucky"#input('inter old brand  :')
new_brand = "Power"#input('inter new brand  :')


with open(file_path, 'r') as j:
    lins = j.readlines()
    filtered_data = []
    for lin in lins:
        data = json.loads(lin)
        if old_word in data['text']:
            # print('')
            data['text'] = replace_text(data['text'], old_word, new_word)
            #if old_brand in data['label']:
            data['label'] = replace_label(data['label'], old_brand, new_brand)
            filtered_data.append(data)

with open('Lucky_Power.jsonl', 'w') as f:
    json.dump(filtered_data, f, ensure_ascii=False)

# Print the number of dictionaries that were modified
print(f"Dictionaries having {old_word} is replaced by {new_word} are {len(filtered_data)}")
```

```python
import json

# Specify the paths to your two input JSONL files
file1_path = '/content/drive/MyDrive/Joyn_NewDataNLP/NER/all_NER_iter1.jsonl'
file2_path = '/content/pura_data.jsonl'

# Specify the path to your output JSONL file
output_file_path = '/content/drive/MyDrive/Joyn_NewDataNLP/NER/all_NER_Argumented.jsonl'

# Read data from the first file
with open(file1_path, 'r', encoding='utf-8') as file1:
    data1 = [json.loads(line) for line in file1.readlines()]

# Read data from the second file
with open(file2_path, 'r', encoding='utf-8') as file2:
    data2 = [json.loads(line) for line in file2.readlines()]

# Combine data from both files
combined_data = data1 + data2
print(len(combined_data))

# Write the combined data to the output file
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    for data in combined_data:
        json_line = json.dumps(data, ensure_ascii=False)
        output_file.write(json_line + '\n')
```
