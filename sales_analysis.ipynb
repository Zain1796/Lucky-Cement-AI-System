```python
def sales_analysis(transcription_text):

    prompt_template = """
    Generate a comprehensive sales analysis report for the cement market in Pakistan, based on conversations with DSR(Companey representative) and outlet owners. The report should focus on all major cement brands and provide insights from a management perspective. The analysis should be approximately 5000 words and include the following sections:


    Sales Performance Analysis:
       Analyze recent sales trends across all major cement brands including Lucky,DG,Falcon,Power,PakLand etc highlighting sales volume, revenue, and market share dynamics. Discuss any notable seasonal trends or shifts in customer demand.

    Customer Insights:
       Summarize feedback from outlet owner regarding customer preferences for different cement brands. Identify key factors influencing purchasing decisions, such as price, quality, brand loyalty, and availability. Discuss preferred payment methods and any logistical challenges faced by customers.

    Competitive Landscape:
       Evaluate the competitive positioning of major cement brands. Highlight strengths and weaknesses as perceived by outlet owner. Compare product offerings, pricing strategies, and customer service levels among leading brands.

    Market Dynamics:
      Explore broader market trends affecting the cement industry, including economic conditions, construction demand, and government regulations. Identify growth opportunities and potential challenges.

    Strategic Recommendations:
      Provide actionable recommendations for all cement brands based on insights gathered. Suggest strategies to enhance sales performance, improve customer satisfaction, and strengthen market positioning.

    Conclusion:
       Summarize the key findings and their implications for the future of the cement market in Pakistan.

    Format Requirements:
       Organize the report with clear headings for each section.
       Include relevant quotes and data from the conversations to support the analysis.
       Maintain a professional and analytical tone throughout the report.

    Note:
       1)do not mention adational information outside of the data
       2)keep in mind this data is of the past two weeks do anlysis acordingly
"""


    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": prompt_template},
                {"role": "user", "content": 'combined_summaries'}
            ],
            max_tokens=2000,
            temperature=0.05,
        )

        result =  response['choices'][0]['message']['content']


        return result
    except openai.error.InvalidRequestError as e:
        print(f"Invalid request error: {e}")
        return None
    except openai.error.OpenAIError as e:
        print(f"OpenAI API error: {e}")
        return None
```
